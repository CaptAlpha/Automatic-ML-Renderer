```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import tarfile
import urllib
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
df=pd.read_csv(r'C:\Users\ritod\Desktop\Kaam\ML\VIIT_Team_MNIST\Ritodeep\train_.csv',encoding='latin-1')
df.head()
```

```python
# Identifying the categotical columns and label encoding them
le = LabelEncoder()
for col in df:
    if(df[col].dtype=='object'):
        df[col]=le.fit_transform(df[col])
        
    
```

```python
df.head()
```

```python
#Enter the target variable
tar = "click"
a=df.pop(tar)
df[tar]=a
```

```python
df.head()
```

```python
# Identifying the columns with null values and filling them with mean
for col in df:
    if(df[col].isnull().sum()!=0):
        df[col]=df[col].fillna(df[col].dropna().mean())
```

```python
print(df.isnull().sum())
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
sc=StandardScaler()
x_train[:,:]=sc.fit_transform(x_train[:,:])
x_test[:,:]=sc.fit_transform(x_test[:,:])
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
score={}
model1 = LogisticRegression()
model1.fit(x_train, y_train)
y_pred1 = model1.predict(x_test)
score['LogisticRegression()']=(accuracy_score(y_test, y_pred1))
```

```python
model2 = DecisionTreeClassifier(criterion = 'entropy', random_state =0 )
model2.fit(x_train,y_train)
y_pred2 = model2.predict(x_test)
score['DecisionTreeClassifier']=(accuracy_score(y_test, y_pred2))

```

```python
model3=RandomForestClassifier()
model3.fit(x_train,y_train)
y_pred3 = model3.predict(x_test)
score['RandomForestClassifier()']=(accuracy_score(y_test, y_pred3))
```
